# Comparison of the Documents with a 95% threshold

```python
def calculate_document_coverage_with_chunks(source_ids, target_ids):
    """
    Calculate coverage accounting for duplicate IDs (document chunks)
    
    Args:
        source_ids (list): List of source document IDs (with duplicates)
        target_ids (list): List of target document IDs (with duplicates)
    """
    # Count frequency of IDs in both lists
    source_counts = {}
    for id in source_ids:
        source_counts[id] = source_counts.get(id, 0) + 1
        
    target_counts = {}
    for id in target_ids:
        target_counts[id] = target_counts.get(id, 0) + 1
    
    # Calculate matches considering frequency
    total_chunks = len(source_ids)  # Total number of chunks in source
    matching_chunks = 0
    
    # Track detailed matches
    chunk_matches = {}
    missing_chunks = {}
    
    for doc_id, source_count in source_counts.items():
        target_count = target_counts.get(doc_id, 0)
        matches = min(source_count, target_count)  # Number of matching chunks
        matching_chunks += matches
        
        # Track detailed matching info
        chunk_matches[doc_id] = {
            'source_chunks': source_count,
            'target_chunks': target_count,
            'matching_chunks': matches
        }
        
        # Track missing chunks
        if matches < source_count:
            missing_chunks[doc_id] = source_count - matches
    
    # Calculate coverage percentage
    coverage_percentage = (matching_chunks / total_chunks) * 100 if total_chunks > 0 else 0
    
    metrics = {
        'total_source_chunks': total_chunks,
        'total_target_chunks': len(target_ids),
        'unique_source_docs': len(source_counts),
        'unique_target_docs': len(target_counts),
        'matching_chunks': matching_chunks,
        'coverage_percentage': round(coverage_percentage, 2),
        'meets_95_percent_threshold': coverage_percentage >= 95,
        'chunk_details': chunk_matches,
        'missing_chunk_details': missing_chunks
    }
    
    return coverage_percentage, coverage_percentage >= 95, metrics

# Example usage:


coverage, meets_threshold, metrics = calculate_document_coverage_with_chunks(source_doc_ids, target_doc_ids)

# Print detailed results
print(f"Coverage Analysis Results:")
print(f"Coverage Percentage: {metrics['coverage_percentage']}%")
print(f"Meets 95% Threshold: {metrics['meets_95_percent_threshold']}")
print(f"\nChunk Metrics:")
print(f"Total Source Chunks: {metrics['total_source_chunks']}")
print(f"Total Target Chunks: {metrics['total_target_chunks']}")
print(f"Unique Source Documents: {metrics['unique_source_docs']}")
print(f"Unique Target Documents: {metrics['unique_target_docs']}")
print(f"Matching Chunks: {metrics['matching_chunks']}")

print("\nDetailed Chunk Matching:")
for doc_id, details in metrics['chunk_details'].items():
    print(f"\nDocument {doc_id}:")
    print(f"  Source chunks: {details['source_chunks']}")
    print(f"  Target chunks: {details['target_chunks']}")
    print(f"  Matching chunks: {details['matching_chunks']}")

if metrics['missing_chunk_details']:
    print("\nMissing Chunks:")
    for doc_id, missing_count in metrics['missing_chunk_details'].items():
        print(f"Document {doc_id}: Missing {missing_count} chunks")
```

```python
def calculate_coverage_stats(source_ids, target_ids):
    """
    Calculate both document-level and chunk-level coverage statistics
    """
    # Document level analysis (unique documents)
    unique_source_docs = set(source_ids)
    unique_target_docs = set(target_ids)
    matching_docs = unique_source_docs.intersection(unique_target_docs)
    
    # Document level percentages
    doc_coverage = (len(matching_docs) / len(unique_source_docs)) * 100 if unique_source_docs else 0
    
    # Chunk level analysis
    source_counts = {}
    target_counts = {}
    
    # Count chunks for each document
    for id in source_ids:
        source_counts[id] = source_counts.get(id, 0) + 1
    for id in target_ids:
        target_counts[id] = target_counts.get(id, 0) + 1
    
    # Calculate matching chunks
    total_matching_chunks = 0
    chunk_matches = {}
    
    for doc_id in source_counts:
        if doc_id in target_counts:
            # For each document, matching chunks is minimum of source and target counts
            chunk_matches[doc_id] = min(source_counts[doc_id], target_counts[doc_id])
            total_matching_chunks += chunk_matches[doc_id]
    
    # Chunk level percentage
    chunk_coverage = (total_matching_chunks / len(source_ids)) * 100 if source_ids else 0
    
    metrics = {
        # Document level metrics
        'total_unique_source_docs': len(unique_source_docs),
        'total_unique_target_docs': len(unique_target_docs),
        'matching_unique_docs': len(matching_docs),
        'document_coverage_percentage': round(doc_coverage, 2),
        
        # Chunk level metrics
        'total_source_chunks': len(source_ids),
        'total_target_chunks': len(target_ids),
        'total_matching_chunks': total_matching_chunks,
        'chunk_coverage_percentage': round(chunk_coverage, 2),
        
        # Detailed matches
        'chunk_matches_by_doc': chunk_matches,
        'missing_docs': list(unique_source_docs - unique_target_docs)
    }
    
    return metrics

# Example usage with test data
# source_ids = [101, 101, 101, 102, 102, 103]  # 6 chunks, 3 unique docs
# target_ids = [101, 101, 102, 102, 102, 104]  # 6 chunks, 3 unique docs

metrics = calculate_coverage_stats(source_doc_ids, target_doc_ids)

print("Document Level Statistics:")
print(f"Total Unique Source Documents: {metrics['total_unique_source_docs']}")
print(f"Total Unique Target Documents: {metrics['total_unique_target_docs']}")
print(f"Matching Unique Documents: {metrics['matching_unique_docs']}")
print(f"Document Coverage: {metrics['document_coverage_percentage']}%")
print(f"Missing Documents: {metrics['missing_docs']}")

print("\nChunk Level Statistics:")
print(f"Total Source Chunks: {metrics['total_source_chunks']}")
print(f"Total Target Chunks: {metrics['total_target_chunks']}")
print(f"Total Matching Chunks: {metrics['total_matching_chunks']}")
print(f"Chunk Coverage: {metrics['chunk_coverage_percentage']}%")

print("\nDetailed Chunk Matches by Document:")
for doc_id, matches in metrics['chunk_matches_by_doc'].items():
    print(f"Document {doc_id}:")
    print(f"  Source chunks: {source_doc_ids.count(doc_id)}")
    print(f"  Target chunks: {target_doc_ids.count(doc_id) if doc_id in target_doc_ids else 0}")
    print(f"  Matching chunks: {matches}")
```
